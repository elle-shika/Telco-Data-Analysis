```python
churn_data = pd.read_csv('telco_churn-230301-080920.csv')
churn_data.head(3)
```

 ## Question 1

```python
#total customer charge = day charge + evening charge + international charge
churn_data['total_customer_charge'] = churn_data['Total day charge'] + churn_data['Total eve charge'] + churn_data['Total intl charge']
churn_data.head(2)
# total customer charge is NaN due to missing values or undefined values


#handling missing variables
#drop new column
churn_data.drop( columns = 'total_customer_charge', inplace = True)
#check for missing values
churn_test = churn_data.isnull()
churn_test.head()
```

```python
#find columns with missing values
churn_data.isnull().sum()
```

```python
#fill in the missing values with pad method
churn_clean = churn_data.bfill()
churn_clean.head()

#create new column total customer charge
churn_clean['total customer charge'] = churn_clean['Total day charge'] + churn_clean['Total eve charge'] + churn_clean['Total intl charge']
churn_clean.head()
```

 #### 2. which customer received the highest number of voicemails
 #### which state do they belong

```python
highest_voice = max(churn_clean['Number vmail messages'])
customer = churn_clean[churn_clean['Number vmail messages'] == highest_voice]

customer_index = customer.index[0]
customer_index

customer_state = customer['State'].values[0]
customer_state
print(f' Customer No.{customer_index} received the highest number of calls and belongs to the {customer_state} state.')
```

 ## 3. Customers with international plan with international charge greater than $3.7

```python
intl_customers = churn_clean[(churn_clean['International plan'] == 'Yes') & (churn_clean['Total intl charge'] > 3.7)]
customer_count = intl_customers['State'].count()
print(f'{customer_count} customers have an international plan with international charge greater than $3.7.')
```

## 4. Number of customers who churned

```python
churn_ed = churn_clean[churn_clean['Churn'] == True]
churned_num = churn_ed['Churn'].count()
print(f'{churned_num} customers churned.')
churn_ed
```

 ## 5. Compare the number of calls made in the day versus in the night and present the time of the day with the most calls measured in hours

```python
day_calls = churn_clean['Total day calls'].sum()
night_calls = churn_clean['Total night calls'].sum()
if day_calls > night_calls:
    hours = (churn_clean['Total day calls']*churn_clean['Total day minutes']).sum()/60
    print(f'Day time has the most calls of {round(hours, 2)} hours.')
else:
    hours = (churn_clean['Total night calls']*churn_clean['Total night minutes']).sum()/60
    print(f'Night time has the most calls of {round(hours, 2)} hours.')
    
```

 ## 6. How many customers are in Arizona State? 

```python
state_customers = churn_clean['State'].groupby(churn_clean['State']).count() #or churn_clean['State'].value_counts()
state_customers #(gives you a list of each state and the number of coustomers residing in them.
# OR
az_customers = churn_clean[churn_clean['State'] == 'AZ']['State'].count()  #gives a count of customers in the arizona state
print(f'{az_customers} customers reside in Arizona state.')
```

 ## 7. Number of customers in New Jersey who had voicemail and an international plan

```python
New_Jersey_cust = churn_clean[churn_clean['State'] == 'NJ']
New_Jersey_voice = New_Jersey_cust[New_Jersey_cust['Voice mail plan'] == 'Yes']
New_Jersey_intl = New_Jersey_voice[New_Jersey_voice['International plan'] == 'Yes']

Customer_num = New_Jersey_intl['State'].count() #gives a count of the number of rows in the specified column of the data frame
'''Customer_num = New_Jersey_intl['State'].value_counts(), gives a count of the individual unique items in the specified column of the dataframe 
and is the same as grouping and then counting the number of entries per group''' 

print(f'{Customer_num} customers in New Jersey had voicemail and international plan.')
```
